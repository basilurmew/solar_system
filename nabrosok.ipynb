# Изображение орбит с учетом масштаба

```python
# Расстояния от Солнца до планет 

d_mer = 0.58 * 10**8
d_ven = 1.08 * 10**8
d_ear = 1.495 * 10**8
d_mar = 2.28 * 10**8
d_jup = 7.78 * 10**8
d_sat = 14.3 * 10**8
d_urr = 28 * 10**8
d_nep = 45.5 * 10**8
```

*^ Добавлены точные размеры*

```python
import numpy as np
```

```python
# Массив углов
t = np.linspace(0, 2*np.pi, 200)
```

```python
# Уравнения орбит планет

r_sun = 0.007 * 10**8

or_mer = r_sun + d_mer * np.transpose([np.cos(t), np.sin(t)])
or_ven = r_sun + d_ven * np.transpose([np.cos(t), np.sin(t)])
or_ear = r_sun + d_ear * np.transpose([np.cos(t), np.sin(t)])
or_mar = r_sun + d_mar * np.transpose([np.cos(t), np.sin(t)])
or_jup = r_sun + d_jup * np.transpose([np.cos(t), np.sin(t)])
or_sat = r_sun + d_sat * np.transpose([np.cos(t), np.sin(t)])
or_urr = r_sun + d_urr * np.transpose([np.cos(t), np.sin(t)])
or_nep = r_sun + d_nep * np.transpose([np.cos(t), np.sin(t)])
```

*^ Добавлен радиус солнца*

```python
import matplotlib.pyplot as plt
```

```python
# Изображение орбит

plt.plot(or_mer[:,0], or_mer[:,1])
plt.plot(or_ven[:,0], or_ven[:,1])
plt.plot(or_ear[:,0], or_ear[:,1])
plt.plot(or_mar[:,0], or_mar[:,1])
plt.plot(or_jup[:,0], or_jup[:,1])
plt.plot(or_sat[:,0], or_sat[:,1])
plt.plot(or_urr[:,0], or_urr[:,1])
plt.plot(or_nep[:,0], or_nep[:,1])
#plt.axis([-3 * 10**8, 3 * 10**8, -3 * 10**8, 3 * 10**8])
plt.axis('equal')
```

# Изображение планет в виде точек

```python
# Координаты начального положения планет

sun = [0,0]
mer = [r_sun + d_mer, 0]
ven = [r_sun + d_ven, 0]
ear = [r_sun + d_ear, 0]
mar = [r_sun + d_mar, 0]
```

```python
# Изображение орбит
plt.plot(or_mer[:,0], or_mer[:,1], color = 'black')
plt.plot(or_ven[:,0], or_ven[:,1], color = 'black')
plt.plot(or_ear[:,0], or_ear[:,1], color = 'black')
plt.plot(or_mar[:,0], or_mar[:,1], color = 'black')

# Изображение границ Солнца
sun_border = r_sun * np.transpose([np.cos(t), np.sin(t)])
plt.plot(sun_border[:,0], sun_border[:,1], 'r--')

# Изображение Солнца и планет
plt.scatter(sun[0], sun[1], color = 'orange')
plt.scatter(mer[0], mer[1], color = 'grey')
plt.scatter(ven[0], ven[1], color = 'yellow')
plt.scatter(ear[0], ear[1], color = 'green')
plt.scatter(mar[0], mar[1], color = 'red')

plt.axis('equal')
```

# Анимация движения Марса

```python
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib widget
```

```python
# Массив координат Марса 

year_mar = 687
t_mar = np.linspace(0, 2*np.pi, year_mar)
anim_mar = r_sun + d_mar * np.transpose([np.cos(t_mar), np.sin(t_mar)])
```

```python
fig = plt.figure()
ax = plt.axes()

def anim(frame):
    
    ax.clear()
    
    #Изображение орбит
    ax.plot(or_mer[:,0], or_mer[:,1], color = 'black')
    ax.plot(or_ven[:,0], or_ven[:,1], color = 'black')
    ax.plot(or_ear[:,0], or_ear[:,1], color = 'black')
    ax.plot(or_mar[:,0], or_mar[:,1], color = 'black')
    
    #Изображение Солнца планет
    ax.scatter(sun[0], sun[1], color = 'orange')
    ax.scatter(mer[0], mer[1], color = 'grey')
    ax.scatter(ven[0], ven[1], color = 'yellow')
    ax.scatter(ear[0], ear[1], color = 'green')
    
    #Анимация движения Марса
    ax.plot(anim_mar[frame,0], anim_mar[frame,1], 'ro')
    
    return None


plt.axis('equal')
FuncAnimation(fig, anim, frames=year_mar, interval=100, repeat=False)
```

# Анимация движения планет

```python
# Массивы координат Меркурия
year_mer = 88
t_mer = np.linspace(0, 2*np.pi, year_mer)
anim_mer = r_sun + d_mer * np.transpose([np.cos(t_mer), np.sin(t_mer)])

# Массивы координат Меркурия
year_ven = 225
t_ven = np.linspace(0, 2*np.pi, year_ven)
anim_ven = r_sun + d_ven * np.transpose([np.cos(t_ven), np.sin(t_ven)])

# Массивы координат Меркурия
year_ear = 365
t_ear = np.linspace(0, 2*np.pi, year_ear)
anim_ear = r_sun + d_ear * np.transpose([np.cos(t_ear), np.sin(t_ear)])
```

```python
fig = plt.figure()
ax = plt.axes()

def anim(frame):
    
    ax.clear()
    
    #Изображение орбит
    ax.plot(or_mer[:,0], or_mer[:,1], color = 'black')
    ax.plot(or_ven[:,0], or_ven[:,1], color = 'black')
    ax.plot(or_ear[:,0], or_ear[:,1], color = 'black')
    ax.plot(or_mar[:,0], or_mar[:,1], color = 'black')
    
    #Изображение Солнца
    ax.scatter(sun[0], sun[1], color = 'orange')
    
    #Расчет координат каждой планеты на каждом кадре анимации
    t1 = frame % year_mer
    t2 = frame % year_ven
    t3 = frame % year_ear
    t4 = frame % year_mar
    
    #Анимация движения планет
    ax.plot(anim_mer[t1,0], anim_mer[t1,1], 'go')
    ax.plot(anim_ven[t2,0], anim_ven[t2,1], 'yo')
    ax.plot(anim_ear[t3,0], anim_ear[t3,1], 'go')
    ax.plot(anim_mar[t4,0], anim_mar[t4,1], 'ro')
    
    return None


plt.axis('equal')
FuncAnimation(fig, anim, frames=year_mar, interval=100, repeat=False)
```

# Анимация в 3D

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

def anim(frame):
    
    ax.clear()
    
    #Изображение орбит
    ax.plot3D(or_mer[:,0], or_mer[:,1], 0, color = 'black')
    ax.plot3D(or_ven[:,0], or_ven[:,1], 0, color = 'black')
    ax.plot3D(or_ear[:,0], or_ear[:,1], 0, color = 'black')
    ax.plot3D(or_mar[:,0], or_mar[:,1], 0, color = 'black')
    
    #Изображение Солнца
    ax.scatter(sun[0], sun[1], 0, color = 'orange', label = 'Sun')
    
    #Расчет координат каждой планеты на каждом кадре анимации
    t1 = frame % year_mer
    t2 = frame % year_ven
    t3 = frame % year_ear
    t4 = frame % year_mar
    
    #Анимация движения планет
    ax.plot3D(anim_mer[t1,0], anim_mer[t1,1], 0, color = 'grey', marker = 'o', label = 'Mercury')
    ax.plot3D(anim_ven[t2,0], anim_ven[t2,1], 0, 'yo', label = 'Venus')
    ax.plot3D(anim_ear[t3,0], anim_ear[t3,1], 0, 'go', label = 'Earth')
    ax.plot3D(anim_mar[t4,0], anim_mar[t4,1], 0, 'ro', label = 'Mars')
    
    plt.legend()
    
    return None


FuncAnimation(fig, anim, frames=year_mar, interval=100, repeat=False)
```

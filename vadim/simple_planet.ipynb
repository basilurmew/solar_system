```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

 # функция для вращения сферы на заданный угол
def rotate(angle):
    ax.view_init(azim=angle)

# создание 3D-графика и сферы
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
u, v = np.mgrid[0:2*np.pi:40j, 0:np.pi:40j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
ax.plot_surface(x[:11], y[:11], z[:11], color='red')
ax.plot_surface(x[10:], y[10:], z[10:], color='blue')
ax.set_axis_off()
ax.set_xticks([])
ax.set_yticks([])
print(x)
plt.show()
# создание анимации
ani = FuncAnimation(fig, rotate, frames=np.linspace(0, 360, 360), repeat = False, interval=5)
# отображение анимации
```

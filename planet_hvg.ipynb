```python
import numpy as np
import math
```

```python
from matplotlib import pyplot as plt
```

```python
ocr = np.linspace(0,2*math.pi, 360)
```

```python
def ell(exc, a):
    b = ((exc*exc - 1)*(-a*a))**0.5
    return a,b
```

```python
merc = ell(0.20563593, 0.38709927)
```

```python
merc_orbit = []
```

```python
merc_per = 0.30749951
```

```python
for i in range(0,len(ocr)):
    merc_orbit.append([merc[0]*math.cos(ocr[i]),merc[1]*math.sin(ocr[i])])
merc_orbit  = np.array(merc_orbit) - np.array([merc[0]-merc_per, 0])
```

```python
plt.plot(merc_orbit[:,0], merc_orbit[:,1])
plt.plot(venera.orbit[:,0],venera.orbit[:,1])
plt.plot(mars.orbit[:,0],mars.orbit[:,1])
plt.plot(zemlya.orbit[:,0],zemlya.orbit[:,1])
plt.scatter(0,0, s = 100, c = 'y', marker = '*')
plt.axis('equal')
```

```python
class planet():
    def __init__(self, exc, a, pereg):
        self.exc = exc
        self.a = a
        self.b = ((exc*exc - 1)*(-a*a))**0.5
        self.pereg = pereg
        self.orbit = []
    def orbita(self):
        
        for i in range(0,len(ocr)):
            self.orbit.append([self.a*math.cos(ocr[i]),self.b*math.sin(ocr[i])])
        self.orbit  = np.array(self.orbit) - np.array([self.a - self.pereg, 0])
        
```

```python
mars = planet(0.0933941,1.523662, 1.381)
```

```python
mars.orbita()
```

```python
venera = planet(0.0068, 0.723332, 0.71843270)
```

```python
venera.orbita()
```

```python
zemlya = planet(0.01671123, 1.00000261, 0.98329134)
zemlya.orbita()
```

```python
jup = planet(0.048775, 5.204267, 4.950429 )
jup.orbita()
```

```python
saturn = planet(0.055723219, 9.58 ,9.048)
saturn.orbita()
```
